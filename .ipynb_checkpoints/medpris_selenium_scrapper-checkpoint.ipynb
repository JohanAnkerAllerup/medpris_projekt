```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time

options = webdriver.ChromeOptions()
driver = webdriver.Chrome("/Users/johanankerallerup/Downloads/chromedriver", chrome_options=options)
driver.get('https://www.medicinpriser.dk/Default.aspx?id=15&vnr=466094')
time.sleep(3)

tab_button = driver.find_element_by_id('ctl00_TopBar_Pridudvikling')
tab_button.click()
time.sleep(3)

firstday = driver.find_element_by_xpath("""//*[@id="ctl00_ctl07_ctl00_DateFrom"]""")
lastday = driver.find_element_by_xpath("""//*[@id="ctl00_ctl07_ctl00_DateTo"]""")
time.sleep(2)

lastday.clear()
firstday.clear()

firstday.send_keys("01.01.2019")
lastday.send_keys("01.01.2020")


refresh = driver.find_element_by_xpath("""//*[@id="ctl00_ctl07_ctl00_CompareButton"]""")
refresh.click()


html = driver.page_source
time.sleep(2)
print(html)
```
